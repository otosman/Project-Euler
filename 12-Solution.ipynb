# find the first triangle number with 500 divisors

```python
import numpy as np

#gets # of divisors
def calcFactors(n):
    if n % 2 == 0:
        n = n/2
    count = 1 
    alpha = 0
    while(n%2 == 0):
        alpha += 1
        n = n/2
    count *= (1+alpha)
    
    nextPrime = 3 
    
    while n != 1:
        alpha = 0
        while n % nextPrime == 0:
            alpha += 1
        count *= (1+alpha)
        nextPrime += 2
    return count

def countTriFact(num):
    count = 1
    factors_n = calcFactors(count)
    factors_n1 = calcFactors(count+1)
    
    while (factors_n*factors_n1) < num:
        count+=1
        factors_n =factors_n1 
        factors_n1 = calcFactors(count+1)
    return count

countTriFact(500)
    
```

```python
#getting the nth triangular number
n = countTriFact(500)

#the actual number is
num = (n*(n+1))/2
```
